```python
#导入训练集和测试集
import pandas as pd
import numpy as np
# 使用read_csv()方法进行读取
train = pd.read_csv(r'C:/Users/LENOVO/Downloads/train (2).csv',index_col=0)
test  = pd.read_csv(r'C:/Users/LENOVO/Downloads/test (2).csv',index_col=0)
train.head()
```

```python
#因为存在很多列存在缺省值，测试集和训练集都有
#如果测试集没有处理，则无法使用模型进行预测
#所以需要将测试集和训练集合并处理
#但是测试集中没有房价，所以需要首先处理训练集的房价
#先对训练集房价进行观察，显示房价价格的条形
%matplotlib  inline
prices=pd.DataFrame({"prices":train["SalePrice"]})
prices.hist()
```

```python
#因为显示房价的条形图并不符合正态，所以我将其变为正态，使之适合机器学习,更加准确
%matplotlib  inline
prices=pd.DataFrame({"log(prices+1)":np.log1p(train["SalePrice"])})
prices.hist()
#因为本身不平滑，所以使用log让他平滑，最后需要变化回原本数据
```

```python
#删除SalePrice这一列
y_train = np.log1p(train.pop('SalePrice'))
#合并两个数据集
all_train=pd.concat((train,test),axis=0)
all_train.head()
```

```python
#变量转化，因为房价中有些数据是文字，属于category变量，
#我需要将其转化为数字才能进行数据分析
#MSSubClass中的数据没有大小关系
all_train['MSSubClass'] = all_train['MSSubClass'].astype(str)
all_train['MSSubClass'].value_counts()
```

```python
#把category的变量转变成numerical表达形式
#使用pandas自带的get_dummies方法实现一键one_hot
pd.get_dummies(all_train['MSSubClass'], prefix='MSSubClass').head()
```

```python
#同理，我将所有category转化
all_dummy_train=pd.get_dummies(all_train)
all_dummy_train.head()
```

```python
#因为部分数据存在缺省值，实现缺省值填充
all_dummy_train.isnull().sum().sort_values(ascending=False).head(10)
```

```python
mean_cols = all_dummy_train.mean()
mean_cols.head(10)
```

```python
all_dummy_train=all_dummy_train.fillna(mean_cols)
```

```python
all_dummy_train.isnull().sum().sum()
```

```python
#标准化numerical数据
#识别numerical数据
numeric_cols = all_train.columns[all_train.dtypes != 'object']
numeric_cols
```

```python
#实现数据点的平滑
numeric_col_means = all_dummy_train.loc[:, numeric_cols].mean()
numeric_col_std = all_dummy_train.loc[:, numeric_cols].std()
all_dummy_train.loc[:, numeric_cols] = (all_dummy_train.loc[:, numeric_cols] - numeric_col_means) / numeric_col_std
```

```python
#建立模型
#将数据集分回训练集和测试集
dummy_train_df = all_dummy_train.loc[train.index]
dummy_test_df = all_dummy_train.loc[test.index]
```

```python
dummy_train_df.shape, dummy_test_df.shape,y_train.shape
```

```python
#我使用Ridge Regression 模型，该模型适用于多因子的数据集，刚好符合该数据集
#将DF转化成Numpy Array，便于符合Sklean
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
X_train = dummy_train_df.values
X_test = dummy_test_df.values
```

```python
alphas = np.logspace(-3, 2, 50)
test_scores = []
for alpha in alphas:
    clf = Ridge(alpha)
    test_score = np.sqrt(-cross_val_score(clf, X_train, y_train, cv=10, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(alphas, test_scores)
plt.title("Alpha vs CV Error");
#可见，大概alpha=10~20的时候，可以把score达到0.135左右。
```

```python
from sklearn.ensemble import RandomForestRegressor
max_features = [.1, .3, .5, .7, .9, .99]
test_scores = []
for max_feat in max_features:
    clf = RandomForestRegressor(n_estimators=200, max_features=max_feat)
    test_score = np.sqrt(-cross_val_score(clf, X_train, y_train, cv=5, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
plt.plot(max_features, test_scores)
plt.title("Max Features vs CV Error");
```

```python
#我们把最好的parameter拿出来，做成我们最终的model
ridge = Ridge(alpha=15)
rf = RandomForestRegressor(n_estimators=500, max_features=.3)
```

```python
ridge.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
#上面提到了，因为最前面我们给label做了个log(1+x), 
#于是这里我们需要把predit的值给exp回去，并且减掉那个"1"
#所以就是我们的expm1()函数。
y_ridge = np.expm1(ridge.predict(X_test))
y_rf = np.expm1(rf.predict(X_test))
```

```python
y_final = (y_ridge + y_rf) / 2
y_final = (y_ridge + y_rf) / 2
submission_df = pd.DataFrame(data= {'Id' : test.index, 'SalePrice': y_final})
```
